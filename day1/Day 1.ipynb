```r
library(foreign) # library for importing sav file (SPSS) into R.
```

```r
dataset = read.spss("Salt.sav", to.data.frame=TRUE)
```

```r
# Look at the dataset
```

```r
str(dataset)
```

```r
plot(dataset$X, dataset$Y)
```

```r
# Q1: Plot the scatter diagram of reduction in triglyceride against weight-loss, add the regression line and comment.
```

```r
fitted_model <- lm(Y ~ X, dataset)
```

```r
summary(fitted_model)
```

```r
# Y = 55.35833 X + 0.48976 
```

```r
# plot with regression model
plot(dataset$X, dataset$Y, main='Concentration of Salt in blood', xlab='Sodium Concentration  g/ml', ylab = 'adsflkjadsf')
abline(fitted_model, col='red')
```
